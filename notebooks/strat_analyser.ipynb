```python
import pandas as pd
import numpy as np
import random

source_file = '../backtest/data/breakout/results_breakout_2020.csv'
cleaned_file = '../backtest/data/breakout/results_breakout_2020_cleaned.csv'
best_strat_file = '../backtest/data/breakout/best_strats2020.csv'
```

```python
# Nettoie les fichiers de toutes les lignes qui n'ont pas le bon format
with open(source_file, 'r') as src, open(cleaned_file, 'w') as dest:
    for line in src:
        if line.count(',') == 7:
            dest.write(line)
```

```python
df = pd.read_csv(cleaned_file)
```

```python
# Filtrer pour obtenir uniquement les stratégies rentables
strategies_rentables = df[df['perf'] > 0]

# Compter le nombre de stratégies rentables
nombre_strategies_rentables = strategies_rentables.shape[0]
print(f"Nombre de stratégies rentables: {nombre_strategies_rentables}")

# Trier par performance décroissante et obtenir le top 30
top_strategies = strategies_rentables.sort_values(by='perf', ascending=False).head(10000)
print("Top des stratégies rentables par performance:")
print(top_strategies)
```

```python
top_strategies.head(10000).to_csv(best_strat_file)
```
